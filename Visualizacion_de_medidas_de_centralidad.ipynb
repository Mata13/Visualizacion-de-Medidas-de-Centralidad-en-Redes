```python
# Importamos las librerías que vamos a usar
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
# Cargamos la base de datos en Jupyter
ruta = 'mammalia-dolphin-florida-overall.edges'

# Vemos los datos y los guardamos en una variable
datos = pd.read_csv(ruta,
            #nrows = 15, # leemos las primeras 10 líneas
            skiprows = 0, # salta líneas
            header = None, # encabezado
            sep = ' ') # separamos los datos en dos columnas que traen los datos
```

```python
# Convertimos los datos en una red
G = nx.from_pandas_edgelist(datos, source = 0, target = 1) # datos es una lista de enlaces

# Graficamos la red
nx.draw(G)
```

```python
# Graficamos la red anterior nuevamente
plt.figure( figsize = [8,8] ) # para hacer la figura más grande
nx.draw(G, node_size = 150, # tamaño de los nodos
        node_color = 'LightCoral', # color de los nodos, usamos de css name color
        node_shape = '<') # forma de los nodos
```

```python
# Eigenvector Centrality
diccionario = nx.eigenvector_centrality(G)
```

```python
# Hacemos un array de comprehension con el diccionario
sizes = np.array([diccionario[i] for i in G])
```

```python
# Revisamos la centralidad Closeness
diccionario = nx.closeness_centrality(G)
colors = np.array( [diccionario[i] for i in G] ) # el color acepta un arreglo y lo guardamos en un arregl
```

```python
# Generamos una gráfica usando los valores de la gráfica anterior
plt.figure( figsize = [20,25] )
nx.draw(G,
        node_size = sizes * 1000,
        node_color = 'gold')
```

```python
# Hacemos una gráfica de G
plt.figure(figsize = [15,15])
nx.draw(G, node_size = 100, node_color = colors, cmap = 'hot') # color map para matplotlib cmap 
```

```python
# Graficamos
plt.figure(figsize = [12,12])
nx.draw_kamada_kawai(G, node_size = sizes * 10000, node_color = colors, cmap = 'viridis') # nos separa en racimos como un arbol
plt.savefig('ruta_imagen.png', dpi = 300)
```

```python
# Generamos la matriz de adyacencia
A = nx.to_numpy_array(G)

# Dibujamos la matriz de una matriz real
plt.title("Matríz de adyacencia de una red real")
plt.imshow(A)
```

```python
# Degree Centrality
diccionario = nx.degree_centrality(G)
d = np.array( [diccionario[i] for i in G] )

# Centralidad Closeness
diccionario = nx.closeness_centrality(G)
c = np.array( [diccionario[i] for i in G] )

# Betweenness Centrality
diccionario = nx.betweenness_centrality(G)
b = np.array( [diccionario[i] for i in G] )

# Eigenvector Centrality
diccionario = nx.eigenvector_centrality(G)
e = np.array( [diccionario[i] for i in G] )

# Graficamos
plt.figure(figsize = [20,5]) # hacemos un lienzo y graficamos dentro lo de abajo
plt.plot(d, label = 'Degree')
plt.plot(c, label = 'Closeness')
plt.plot(b, label = 'Betweenness')
plt.plot(e, label = 'Eigenvector')
plt.xlabel('Nodos')
plt.ylabel('Centralidades')

plt.title("Centralidades")
plt.legend() # para que aparezcan las legend
```

<font color='DarkSalmon'> Graficamos usando el Clustering Coefficient for nodes (Coeficiente de Acumulación) asociado con el color de cada nodo </font>

```python
# Obtenemos un diccionario de que a cada nodo le da una posición
position = nx.layout.kamada_kawai_layout(G)

# Clustering, a cada nodo le asigna un valor
diccionario = nx.clustering(G)
clus = np.array( [diccionario[i] for i in G] )
```

```python
# Dibujamos la red con el coeficiente de acumulación (Clustering Coefficient for nodes)
plt.figure(figsize = [10,10])
nx.draw(G, pos = position, node_color = clus)
plt.savefig('closter.png', dpi = 300)
plt.title("Mammalia-Dolphin-Florida-Overall network using Clustering Coefficient")
```

```python
# Forzamos a convertir los Grados de G a un diccionario
diccionario = dict(G.degree)

# Guardamos los grados
grados = [diccionario[i] for i in G] # nos deshacemos de las claves

# Vemos a grados
grados
```

```python
# Creamos un diccionario vacio que vamos a llenarno
distribucion = {}

for i in grados:
    distribucion[i] = 0

for i in grados:
    distribucion[i] += 1 # a lo que valía le suma uno

distribucion # el número de nodos que tienen grado x lo tienen y 
```

```python
x = distribucion.keys()
y = distribucion.values()
```

```python
# Graficamos el histograma 
plt.figure(figsize = [15,5])
plt.bar(x, y)
plt.xlabel('Nodos')
plt.ylabel('Grados')

plt.savefig('histogram_network.png', dpi = 600)
plt.title("Mammalia-Dolphin-Florida-Overall network histogram")
plt.legend() # para que aparezcan las legend
```
